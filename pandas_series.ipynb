
# **Indexing, Selection and Slicing**
Individual columns in the DF can be selected with regular indexing. Each columns is represented as a Series.

```python
df
```

```python
df.loc['Canada'] # gives all attributes of Canada ie., accessing rows using index name
```

```python
df.iloc[-1] #iloc used with numbers; -1 gives last index
```

```python
df['Population'] #accessing the columns
```

```python
df['Population'].to_frame() #to get the name and a more dataframe-like format
```

```python
df[['Population','GDP']] #no need of to_frame() if accessing multiple columns
```

```python
df[1:3] #row level slicing; here 1 to 3 ignoring 1 and giving 2,3
```

```python
df.loc['Italy'].to_frame() #or df.loc['Italy'] also works
```

```python
df.loc['France':'Italy'] #using this is better for slicing; ie., with loc and iloc
```

```python
df.loc['France':'Italy','Population'].to_frame()
```

```python
df.loc['France':'Italy',['Population','GDP']]
```

```python
# can also do all of this using iloc[number]
```

# **Pandas Conditional Selection and Modifying DataFrames**

```python
df
```

```python
df['Population']>70 #to get the boolean list
```

```python
df.loc[df['Population']>70] #accessing with the boolean list
#give all the columns for countried with population above 70
```

```python
df.loc[df['Population']>70,'Population'].to_frame()
#give only the population for the countries with population above 70
```

```python
df.loc[df['Population']>70,['Population','GDP']]
#give only the population and GDP for the countries with population above 70
```

# Dropping stuff
Instead of selecting the things you want to access, you can drop the things you don't want to access. Very useful if you want to exclude some things but get everything else.

```python
df.drop('Canada')
#give everything except Canada's details
```

```python
df.drop(['Canada','Japan'])
```

```python
df.drop(columns=['Population','GDP'])
```

```python
#we can also do operations on columns
df[['Population','GDP']]/100
```

# Operations with Series
These work at a column level, broadcasting down the rows

```python
crisis = pd.Series([-1_000_000,-0.3], index=['GDP','HDI'])
crisis
```

```python
df[['GDP','HDI']]
```

```python
df[['GDP','HDI']] + crisis
#subtracting the crisis series from the dataframe
```

# Modifying DataFrames

```python
langs = pd.Series(
    ['French','German','Italian'],
    index=['France','Germany','Italy'],
    name='Language'
)
```

```python
langs.to_frame() # or langs
```

```python
df['Language'] = langs
```

```python
df
```

```python
df['Language'] = 'English'
```

```python
df
```

```python
df.rename(
    columns={
    'HDI':'Human Development Index',
    'Annual Popcorn Consumption':'APC'
    }, index={
     'United States':'USA',
     'United Kingdom':'UK',
     'Argentina':'AR'
    }
)
```
